```python
import matplotlib.pyplot as plt
import numpy as np
import help_functions as hf
```

```python
def gen_1d_mesh(x0,xmax,k) -> list:
    mesh = [0]
    h = x0
    mesh.append(h)
    while(mesh[-1]<xmax):
        h*=k
        mesh.append(mesh[-1]+h)
    mesh.pop()
    mesh.append(xmax)
    return mesh
    
def paint_mesh(x,y,labels:tuple, limx = 0, limy = 0):
    fig,ax = plt.subplots(figsize=(6,6))
    for i in r:
        ax.axvline(x=i, ymin=0, ymax=z[-1])

    for i in z:
        ax.axhline(y=i, xmin=0, xmax=r[-1])

    if limx: ax.set_xlim(limx)
    else: ax.set_xlim([x[0],x[-1]])
    if limy: ax.set_ylim(limy)
    else: ax.set_ylim([y[0],y[-1]])

    ax.set_xlabel(labels[0])
    ax.set_ylabel(labels[1])
    fig.show()

r = gen_1d_mesh(0.5,30,1.)
z = gen_1d_mesh(0.5,30,1.)
paint_mesh(r,z,("r, m", "z, m"))
len(r)*len(r)
```
